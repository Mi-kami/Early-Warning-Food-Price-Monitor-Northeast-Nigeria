###**6.1 Overview**

This results section covers the evaluation of three forecasting models — Prophet, XGBoost, and SARIMA — tested across 13 commodity-state combinations in the Tier 1 evaluation states of Adamawa, Borno, and Yobe. All three models were measured against a Naive baseline, which
simply assumes that next month&#39;s price will equal this month&#39;s price. A model only adds value if it beats this baseline. The central finding is unambiguous: all 13 combinations achieved production-ready accuracy, meaning every model-commodity-state pairing is reliable enough
to be used in a real planning or humanitarian decision-making context.

Accuracy is measured using **MAPE** — **Mean Absolute Percentage Error**. Think of it as the average percentage by which a forecast is wrong. A MAPE of 14.7% means the model's predictions were off by less than 15 naira in every 100 naira of actual price. Lower is better.

| **MAPE Range** | **What It Means**  |
|---|:---|
|Below 20%| Excellent — reliable for direct operational decisions|
|20% – 35%| Production-ready — useful for planning alongside other signals|
|35% – 100%| Directional only — shows whether prices rise or fall, not by how much|
|Above 100%| Not recommended|

###**6.2 How the Three Models Compared**

Prophet was the dominant model across the evaluation, winning 10 out of 13 commodity-state combinations — a win rate of nearly 77%.

|Prophet| 10 of 13| 76.9%| 32.1%|
|XGBoost| 2 of 13| 15.4%| 49.7%|

Prophet performed best because it was designed to do exactly what Nigerian food price data demands: detect long-term price trends and recurring seasonal patterns at the same time. Food prices in Nigeria are shaped by both forces simultaneously — a sustained upward trend driven by
naira depreciation and the 2023 fuel subsidy removal, layered on top of predictable seasonal cycles tied to harvests and dry seasons. Prophet handles this combination naturally. XGBoost and SARIMA were not as well-suited to this structure in a post-subsidy price environment.

###**6.4 Forecast Accuracy Across Horizons**

The models were also tested at three forecast distances: one month ahead (H1), two months ahead (H2), and three months ahead (H3). SARIMA was evaluated at H1 only; the H2 and H3 comparison is between Prophet and XGBoost.

|Forecast Horizon| Prophet Avg MAPE| XGBoost Avg MAPE|
|---|:---:|:---:|
|H1 — 1 month ahead| 24.5%| 51.1%|
|H2 — 2 months ahead| 28.5% |48.3%|
|H3 — 3 months ahead| 35.3%| 49.7%|

Prophet shows the pattern you would expect from a well-functioning forecasting tool: accuracy decreases gradually as the forecast looks further ahead — from 24.5% at one month to 35.3% at three months. The further into the future you look, the more uncertainty accumulates, just as a weather forecast is more reliable for tomorrow than for next week. This gradual increase in error is healthy and expected.

XGBoost tells a different story. Its error barely changed across all three horizons — 51.1% at one month, 48.3% at two months, 49.7% at three months. This flat pattern is not a good sign; it is a diagnostic signal. A model with genuine forecasting ability should become less accurate the
further out it predicts. XGBoost's flat error indicates it was already operating at the edge of its capability at one month ahead — and the reason for this is explained in the next section.

###**6.5 The Impact of the 2023 Fuel Subsidy Removal**

All three models were trained on price data from before June 2023. When the Nigerian government removed the fuel subsidy in June 2023, food prices jumped sharply to levels the models had never encountered in their training data. The result was a structural break — a moment where the historical patterns the models had learned were no longer a reliable guide to what prices would do next.

This is like asking someone to predict the price of petrol using only data from before a price cap was removed. Once the cap lifts and prices leap to a new level, predictions based on the old pattern will systematically underestimate the new reality — not because the model is poorly built, but because the world it was trained on no longer exists.

Prophet handled this disruption better than XGBoost for a specific reason: it is designed to project a trend line forward even into price territory it has never seen before. When prices surged past the training range, Prophet extended its trend estimate forward and partially tracked the new levels. XGBoost cannot do this — it is constrained by its training data range and cannot reliably predict values higher than the highest price it was trained on. This is precisely why XGBoost&#39;s
accuracy was flat and weak at all three forecast horizons: from the very first month of the forecast period, it was already extrapolating beyond its training window, and this structural limitation persisted regardless of how far ahead it was asked to predict.

The subsidy removal was the single biggest external factor affecting these results. It did not invalidate the models — all 13 combinations remained production-ready despite the shock — but it does underline the need to retrain models regularly as new price data becomes available,
particularly following major policy changes.
